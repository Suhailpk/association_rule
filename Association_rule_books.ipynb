## Import necessary libaries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import data

```python
books = pd.read_csv('book (1).csv')
books
```

## Assocaition rules 20%  Support and 80% confidence

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
books_freq_datasets = apriori(books,min_support=0.2,use_colnames=True)
books_freq_datasets
```

```python
association_rule1 = association_rules(books_freq_datasets,metric='confidence',min_threshold=0.8)
association_rule1
```

### Visualization

```python
plt.scatter(x=association_rule1['support'],y=association_rule1['confidence'])
plt.title('Association rules with 20% support and 80% confidence',size=20)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```
